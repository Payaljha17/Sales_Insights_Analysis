```python
#IMPORTING CSV
import pandas as pd
df = pd.read_csv('all_months_data_2025.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
#FINDING NULL VALUES
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)].head(20)
```

```python
#Droping NaN values
df = df.dropna(how='all')
```

```python
#NULL Values in Dataset
df.isnull().sum()
```

```python
#Dropping null values of order_date
df = df.dropna(subset=['Order Date'])
```

```python
#NULL Values in dataset = 0
df.isnull().sum()
```

```python
df.head(5)
```

```python
#FORMATTNG COLUMNS NAME
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
#Adding months Column
df['month'] = df['order_date'].dt.month
```

```python
df.head(5)
```

```python
#COVERTING column to correct datatype
df['quantity_ordered'] = pd.to_numeric(df['quantity_ordered'], errors='coerce')
df['price_each'] = pd.to_numeric(df['price_each'], errors='coerce')

```

```python
#ADDING A SALES COLUMN
df['sales'] = df['quantity_ordered'] * df['price_each']
df.head()
```

```python
#Grouping by months, sales
df.groupby('month')['sales'].sum()
```

```python
#ADDING A CITY COLUMN
df['city'] = df['purchase_address'].apply(lambda x: x.split(',')[1])
df.head()
```

```python
#Plotting Chart
import matplotlib.pyplot as plt

df['month'] = df['month'].astype(int)
monthly_sales = df.groupby('month')['sales'].sum().sort_index()

months = range(1,13)

plt.bar(months, monthly_sales)
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month Number')
plt.title('Monthly Sales')

plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

```python
#Grouping by sales in city
df.groupby('city')['sales'].sum()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group by City
city_results = df.groupby('city')['sales'].sum()
plt.figure(figsize=(12,6))
plt.bar(city_results.index, city_results.values)
plt.xticks(rotation=90)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.title('Sales by City')


plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

```python
df['hour'] = pd.to_datetime(df['order_date']).dt.hour
df['minute'] = pd.to_datetime(df['order_date']).dt.minute
df.head()
```

```python
hours = [hour for hour, df in df.groupby('hour')]

plt.plot(hours, df.groupby(['hour']).count())
plt.xticks(hours)
plt.grid()
plt.show()
```

```python
# WHAT PRODUCTS ARE MOST OFTEN SOLD ?
df.groupby('product')['quantity_ordered'].sum().plot(kind='bar')
```

```python
#EXTRACTING COMMON ORDER ID
temp = df[df['order_id'].duplicated(keep=False)]

#Grouping common id product
temp['grouped'] = temp.groupby('order_id')['product'].transform(lambda x: ','.join(x))

#Deleting Duplicates Column
temp[['order_date', 'grouped']].drop_duplicates()
temp.head()
```
